## Création de la base de données

```
folder_path = "gdrive/MyDrive/ISEN/NLP/Partie_1"

dict_data = {}

for filename in os.listdir(folder_path):
    if filename.endswith(".story"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            content = file.read()
            dict_data[filename] = content

df = pd.DataFrame(list(dict_data.items()), columns=['filename', 'content'])
df.head()
```

## En minuscule

```
df['content'] = df['content'].str.lower()

df.head()
```

## Tokenization

```
df['content'] = df['content'].apply(word_tokenize).apply(lambda liste: [mot for mot in liste if mot.isalpha()])

df.head()
```

## Visualization du vocabulaire

```
print(df['content'][0])
```

## Les stopwords

```
len(nltk.corpus.stopwords.words('english'))
```

```
df['content'] = df['content'].apply(lambda liste: [mot for mot in liste if mot not in nltk.corpus.stopwords.words('english')])

df.head()
```

## Stemming

```
from nltk.stem.snowball import EnglishStemmer

stemmer = EnglishStemmer()

df['content'] = df['content'].apply(lambda liste: [stemmer.stem(mot) for mot in liste])

df.head()
```

## Lemmatization

```
from nltk.stem import WordNetLemmatizer
Word_Lemmatizer = WordNetLemmatizer()

df['content'] = df['content'].apply(lambda token: [Word_Lemmatizer.lemmatize(mot) for mot in token])

df.head()
```

## Importation des données

Les données proviennent de ce [jeux de données](https://zenodo.org/record/841984#.X_Jb2ulKjBI).

```
x_text = pd.read_csv('gdrive/MyDrive/ISEN/NLP/Partie_2/x_text.csv')
y_text = pd.read_csv('gdrive/MyDrive/ISEN/NLP/Partie_2/y_text.csv')
print(x_text.shape)
print(y_text.shape)
```

```
x_text.head()
```

```
y_text.head()
```

```
lang = np.unique(y_text)
print(lang)
```

```
stp_en = nltk.corpus.stopwords.words('english')
stp_fr = nltk.corpus.stopwords.words('french')
stp_it = nltk.corpus.stopwords.words('italian')
stp_es = nltk.corpus.stopwords.words('spanish')
stp_ru = nltk.corpus.stopwords.words('russian')

x_text['lang'] = ""
x_text['valid'] = 0

for text in x_text['text']:
  count_fr = 0
  count_en = 0
  count_it = 0
  count_es = 0
  count_ru = 0
  for word in text.split():
    if word in stp_fr:
      count_fr += 1
    if word in stp_en:
      count_en += 1
    if word in stp_it:
      count_it += 1
    if word in stp_es:
      count_es += 1
    if word in stp_ru:
      count_ru += 1

  if(count_fr > count_en and count_fr > count_it and count_fr > count_es and count_fr > count_ru):
    x_text.loc[x_text['text'] == text, 'lang'] = "french"
  elif(count_en > count_fr and count_en > count_it and count_en > count_es and count_en > count_ru):
    x_text.loc[x_text['text'] == text, 'lang'] = "english"
  elif(count_it > count_fr and count_it > count_en and count_it > count_es and count_it > count_ru):
    x_text.loc[x_text['text'] == text, 'lang'] = "italian"
  elif(count_es > count_fr and count_es > count_en and count_es > count_it and count_es > count_ru):
    x_text.loc[x_text['text'] == text, 'lang'] = "spanish"
  elif(count_ru > count_fr and count_ru > count_en and count_ru > count_it and count_ru > count_es):
    x_text.loc[x_text['text'] == text, 'lang'] = "russian"
  else:
    x_text.loc[x_text['text'] == text, 'lang'] = "other"
```

```
print(x_text.groupby('lang').count())
```

```
# compare label in text_y with lang in text_x

x_text['valid'] = np.where(x_text['lang'] == y_text['label'], 1, 0)
print(x_text.groupby('valid').count())

# Print % error.
print(x_text.groupby('valid').count()['text'][1] / x_text.shape[0] * 100, "%")
```
