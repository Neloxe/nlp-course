# Création de la base de données

```
cmpt = 0
dict_data = dict()

for file_name in os.listdir(path):
    f = open(path+'/'+file_name, 'r')
    lst = ""
    for line in f:

       line.strip()
       line = line.replace("\n" ,'')
       line = line.replace("//" , '')
       line = line.replace("/" , '')
       if len(line) > 0 :
        lst += line

    dict_data[cmpt] = lst
    cmpt += 1


print(dict_data)
```

# Prétraitements

## En minuscule

```
dict_data_min = dict()
for k, v in dict_data.items():
  dict_data_min[k] = v.lower()
```

## Tokenization

```
dict_data_token = dict()
for k, v in dict_data_min.items():
  tokenizer = nltk.RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(v)
  dict_data_token[k] = tokens

print(dict_data_token)
```

## Stopwords

```
dict_data_stop = dict()
for k, v in dict_data_token.items():
  dict_data_stop[k] = [w for w in v if not w in list(nltk.corpus.stopwords.words())]
```

```
print(dict_data_stop)
```

## Stemming

```
dict_data_stem = dict()
st = LancasterStemmer()

for index, doc in dict_data_stop.items():
  dict_data_stem[index] = [st.stem(w) for w in doc ]

print(dict_data_stem)
```

# Bag of words

## Vecteur de vocabulaire

```
vocab = set()

for index, doc in dict_data_stem.items():
  for word in doc :
    vocab.add(word)

print(len(vocab))
print(vocab)
```

## Bag of words

```
vocabulary = list(vocab)
data = pd.DataFrame([], columns=vocabulary)

for index, doc in dict_data_stem.items() :
  # Your Code
  for word in doc :
    if index not in data.index:
      data.loc[index] = 0
    if word in data.columns :
      data.loc[index, word] += 1
    else :
      data.loc[index, word] = 1

data.head()

# Index is the ID of the doc.
```
